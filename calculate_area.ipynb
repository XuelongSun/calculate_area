```python
import numpy as np

# 求角a
angle_a = np.arctan(2.0/4.0)
# 求两个弓形的面积
s1 = (angle_a * 2.0) / (np.pi*2) * (np.pi*4**2) - 2 * (4.0*np.sin(angle_a)*4.0*np.cos(angle_a)/2.0)
s2 = ((np.pi/2 - angle_a) * 2.0) / (np.pi*2) * (np.pi*2**2) - 2 * (2.0*np.sin(angle_a)*2.0*np.cos(angle_a)/2.0)
# 加起来得到阴影部分的面积
s = s1 + s2
# 打印结果
print(s)
```

```python
import numpy as np
# 仍多少硬币
N = 5000
# 随机数 x,y定义了扔出硬币的位置，都在 0-4, 扔硬币要扔到正方形里面
r_x = np.random.uniform(low=0, high=4, size=N)
r_y = np.random.uniform(low=0, high=4, size=N)
# 硬币要落入阴影部分，就是两个圆相交的部分
f1 = r_x**2 + (r_y - 4)**2 < 16
f2 = r_y**2 + (r_x - 2)**2 < 4
# 要同时落入两个圆内，所以用 f1*f2，也就是True*True=1, False*Ture=0, False*False=0
# 统计1的个数，也就是有多少硬币落入了阴影部分，然后求比例再乘以正方形的面积
s = np.sum(f1*f2) / N * 16
# 打印结果，每次的结果都不一样哦，随机的嘛
print(s)
```

```python
import matplotlib.pyplot as plt
import numpy as np

Ns = np.linspace(100, 1000000, 500).astype(np.int32)
mean_num = 10
area = []
for N in Ns:
    temp = 0
    for i in range(mean_num):
        # 随机数 x,y定义了扔出硬币的位置，都在 0-4, 扔硬币要扔到正方形里面
        r_x = np.random.uniform(low=0, high=4, size=N)
        r_y = np.random.uniform(low=0, high=4, size=N)

        f1 = r_x**2 + (r_y - 4)**2 < 16
        f2 = r_y**2 + (r_x - 2)**2 < 4
        temp += np.sum(f1*f2)
    area.append(temp / (10 * N) * 16)
    
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(Ns[1:], area[1:], lw=2, color='skyblue', label='Monte Carlo')
ax.plot([0, Ns[-1]],[3.846956615189259]*2, 
        ls='--', color='gray', lw=2, label="Standard")
ax.grid(1)
ax.legend()
ax.set_xlabel('Times')
ax.set_ylabel('Area')
```

```python
import numpy as np

def f_r1(x):
    return -np.sqrt(16 - x*x) + 4
def f_r2(x):
    return np.sqrt(4-(x-2)**2)

# x 从0开始
x = 0
# 定义矩形的宽度，越小，越准确
dx = 0.01
# 现给dy一个很大的值
dy = np.inf
# 初始化面积为0
area = 0
# 关键点，f2-f1小于0的时候退出
while dy > 0:
    x += dx
    # 两个函数之差为矩形的长
    dy = f2(x + dx) - f1(x + dx)
    # 累加面积
    area += dx * dy
print(area)
```

```python
import sympy as sp

# 求两个圆相交的坐标
x = sp.Symbol('x')
y = sp.Symbol('y')
solved_value = sp.solve([x**2+(y-4)**2-16,(x-2)**2+y**2-4], [x,y])
print(solved_value)
```

```python
# 使用scipy的integrate对象来进行定积分求解
from scipy import integrate
# 定义积分函数就是f2-f1
def f(x):
    return np.sqrt(4-(x-2)**2) + np.sqrt(16 - x*x) - 4
# 求解
v, err = integrate.quad(f, 0, 16/5)
# 打印结果
print('结果是{}，误差为{}'.format(v,err))
```
